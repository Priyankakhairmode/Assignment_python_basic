3. Explain the following termsg
M The difference in np.array, np.asarray and np.asanyarrayX

**Differences in np.array, np.asarray, and np.asanyarray **
These NumPy functions are used to create arrays, but they differ in how they handle existing ndarray objects and subclasses:
np.array(object, dtype=None, copy=True, order='K', subok=False, ndmin=0, like=None):
Always creates a new array object, making a copy of the input data by default (unless copy=False is explicitly set).
Does not preserve subclasses; the output is always a base ndarray.
np.asarray(a, dtype=None, order=None, like=None):
Returns an ndarray object, but does not make a copy if the input a is already an ndarray with the same dtype.
If a is not an ndarray or has a different dtype, a new ndarray (a copy) is created.
Does not preserve subclasses; the output is always a base ndarray.
np.asanyarray(a, dtype=None, order=None, like=None):
Similar to np.asarray, it returns an ndarray object and avoids copying if the input a is already an ndarray with the same dtype.
Key difference: If a is a subclass of ndarray (e.g., a matrix object), np.asanyarray will return the input object itself, preserving its subclass type.

**Differences between Deep Copy and Shallow Copy**
These terms describe how copies of objects, especially those containing nested structures, are made:
Shallow Copy:
Creates a new compound object but does not create copies of the nested objects within it.
Instead, it inserts references to the original nested objects into the new compound object.
Changes to mutable nested objects in either the original or the shallow copy will affect both, as they share the same underlying nested objects.
Example in Python: list.copy() or copy.copy().
Deep Copy:
Creates a completely independent copy of the compound object, including all nested objects recursively.
It constructs new copies of all objects found in the original, ensuring no shared references.
Changes to any part of the deep copy will not affect the original object, and vice-versa.
Example in Python: copy.deepcopy().

 14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib
a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a time-series dataset
dates = pd.date_range(start="2025-01-01", periods=30, freq="D")
temperature = np.random.randint(20, 35, size=30)   # Random temperatures between 20–35
humidity = np.random.randint(40, 80, size=30)      # Random humidity between 40–80

df = pd.DataFrame({
    "Date": dates,
    "Temperature": temperature,
    "Humidity": humidity
})

# Step 2: Plot Temperature and Humidity with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Temperature on left y-axis
ax1.plot(df["Date"], df["Temperature"], color="red", label="Temperature")
ax1.set_xlabel("Date")
ax1.set_ylabel("Temperature (°C)", color="red")
ax1.tick_params(axis="y", labelcolor="red")

# Create second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df["Date"], df["Humidity"], color="blue", label="Humidity")
ax2.set_ylabel("Humidity (%)", color="blue")
ax2.tick_params(axis="y", labelcolor="blue")

# Step 3: Add title
plt.title("Temperature and Humidity Over Time")

# Step 4: Add legends
fig.legend(loc="upper right", bbox_to_anchor=(0.9, 0.9))

# Show plot
plt.show()
```

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# a) Create NumPy array with 1000 samples from a normal distribution
data = np.random.randn(1000)

# b) Plot histogram with 30 bins
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='b')

# Fit a normal distribution to the data
mu, std = norm.fit(data)

# Overlay PDF line plot
pdf = norm.pdf(bins, mu, std)
plt.plot(bins, pdf, 'k', linewidth=2)

# c) Label axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set title
plt.title('Histogram with PDF Overlay')

# Show plot
plt.show()
```

17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate two random arrays
x = np.random.randn(200)
y = np.random.randn(200)

# Determine quadrant for each point
def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Quadrant I'
    elif x < 0 and y >= 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

quadrants = [get_quadrant(xi, yi) for xi, yi in zip(x, y)]

# Create DataFrame for Seaborn
df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Create scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Show legend
plt.legend(title='Quadrants')
plt.axhline(0, color='black', linewidth=1, linestyle='--')
plt.axvline(0, color='black', linewidth=1, linestyle='--')

plt.show()
```

18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'8

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Display plots inline (use output_file("sine_wave.html") if you want a standalone HTML file)
output_notebook()

# Generate sine wave data
x = np.linspace(0, 4*np.pi, 500)   # values from 0 to 4π
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave Function",
           x_axis_label='X-axis',
           y_axis_label='Y-axis',
           width=700, height=400)

# Add line plot
p.line(x, y, line_width=2, color="navy", legend_label="sin(x)")

# Add grid lines (enabled by default, but we can style them)
p.xgrid.grid_line_color = "gray"
p.ygrid.grid_line_color = "gray"
p.xgrid.grid_line_alpha = 0.5
p.ygrid.grid_line_alpha = 0.5

# Show the plot
show(p)
```

19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'8

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.io import output_notebook

# Display plots inline (use output_file("bar_chart.html") for standalone HTML)
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=len(categories))

# Create DataFrame
df = pd.DataFrame({'Category': categories, 'Value': values})

# Convert to ColumnDataSource
source = ColumnDataSource(df)

# Create figure
p = figure(x_range=categories,
           title="Random Categorical Bar Chart",
           x_axis_label="Category",
           y_axis_label="Value",
           width=700, height=400)

# Add bars with color mapping based on values
p.vbar(x='Category', top='Value', width=0.6, source=source,
       fill_color=factor_cmap('Category', palette="Spectral5", factors=categories))

# Add hover tooltips
hover = HoverTool(tooltips=[("Category", "@Category"), ("Value", "@Value")])
p.add_tools(hover)

# Style grid lines
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show plot
show(p)
```

20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'8

```python
import numpy as np
import plotly.graph_objects as go

# Generate random dataset
x = np.arange(0, 50)              # X values from 0 to 49
y = np.random.randn(50).cumsum()  # Random cumulative sum for smoother line

# Create line plot
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x, y=y,
    mode='lines+markers',
    name='Random Data',
    line=dict(color='blue', width=2)
))

# Label axes and set title
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='Index',
    yaxis_title='Value',
    template='plotly_white'
)

# Show plot
fig.show()
```

21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
