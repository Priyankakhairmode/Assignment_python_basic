```python
#21 Write a Python program to train a Decision Tree Classifier with a maximum depth of 3 and compare its
#accuracy with a fully grown tree

from sklearn import tree
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# load dataset
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data
y = iris.target

# split dataset
# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state =1)

# Train Decision Tree woth max_depth = 3
Classifier_depth_3 = DecisionTreeClassifier(max_depth = 3)
Classifier_depth_3.fit(x_train, y_train)

# train a fully grown tree
Classifier_full_grown = DecisionTreeClassifier()
Classifier_full_grown.fit(x_train, y_train)

# predict and calculate accuracy for both trees
y_pred_depth_3 = Classifier_depth_3.predict(x_test)
y_pred_full_grown = Classifier_full_grown.predict(x_test)

# print the accuracy for comparison
print("accuracy score for y_pred_depth_3:", round(accuracy_score(y_test, y_pred_depth_3),2))
print("accuracy score for y_pred_full_grown:", round(accuracy_score(y_test, y_pred_full_grown),2))

#visualization of Decision tree with depth = 3 and fully grown decision tree
plt.figure(figsize= (8,5))
tree.plot_tree(Classifier_depth_3, filled= True)
plt.figure(figsize= (8,5))
tree.plot_tree(Classifier_full_grown, filled= True)

plt.show()
```

```python
#23 Write a Python program to apply feature scaling before training a Decision Tree Classifier and compare its
#accuracy with unscaled data
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# load dataset
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data
y = iris.target

# split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state= 1)

# apply feature scaling
scaler = StandardScaler()
x_train_Scaled = scaler.fit_transform(x_train)
x_test_Scaled = scaler.transform(x_test)

# train decision tree with scaled data
Classifier_scaled = DecisionTreeClassifier()
Classifier_scaled.fit(x_train_Scaled, y_train)
y_predicted_scaled = Classifier_scaled.predict(x_test_Scaled)
accuracy_scaled = accuracy_score(y_test, y_predicted_scaled)

# train decision tree with unscaled data
Classifier_unscaled = DecisionTreeClassifier()
Classifier_unscaled.fit(x_train, y_train)
y_predicted_unscaled = Classifier_unscaled.predict(x_test)
accuracy_unscaled = accuracy_score(y_test, y_predicted_unscaled)
'''Decision Trees are Non-parametric and Based on Rules:

Non-parametric: Unlike models like K-Nearest Neighbors or Support Vector Machines, Decision Trees do not rely on assumptions about the underlying data distribution. This means they are not sensitive to the scale or magnitude of features.
Rule-based: Decision Trees make predictions based on a series of rules derived from the data. These rules are typically in the form of "if-then-else" conditions applied to feature values. For example, a rule might be: "If petal length is greater than 4.8 cm, then classify as Iris-virginica."
How Decision Trees Work:

Finding the Best Split: At each node of the tree, the algorithm searches for the best feature and split point that maximally separates the data points into different classes. It uses impurity measures like Gini impurity or entropy to evaluate the quality of a split.
Recursive Splitting: The process of finding the best split is repeated recursively for each child node until a stopping criterion is met (e.g., maximum depth, minimum number of samples per leaf).
Prediction: To make a prediction for a new data point, the algorithm traverses the tree based on the feature values of the data point, following the rules at each node until it reaches a leaf node. The class assigned to the leaf node is the predicted class for the data point.
Why Scaling Doesn't Matter:

Splitting Based on Relative Values: Decision Trees focus on the relative ordering of feature values rather than their absolute magnitudes. When determining the best split, the algorithm compares the values of a feature within the data points at a particular node. Scaling the features does not change their relative ordering; it only changes their scale.
Thresholds Adjust Accordingly: Even if features are scaled, the decision boundaries or thresholds used in the tree's rules will adjust accordingly. The algorithm will still find the same optimal split points, even if the feature values themselves have been transformed.
Example:

Imagine a decision rule based on petal length: "If petal length > 4.8 cm, then classify as Iris-virginica."

If we scale the petal length by dividing all values by 10, the rule would become: "If scaled petal length > 0.48, then classify as Iris-virginica." The decision boundary adjusts to the scaled values.
The underlying logic of the rule remains the same: if a data point has a relatively large petal length (whether scaled or unscaled), it is more likely to be an Iris-virginica.
In conclusion:

Since Decision Trees rely on the relative ordering of feature values and adjust their decision boundaries to the data's scale, feature scaling generally has no significant impact on their accuracy. This makes them robust to differences in feature ranges and eliminates the need for scaling in most cases. However, there might be minor exceptions or edge cases, and this is why it can sometimes still be useful to confirm this experimentally, as the code you shared does.
'''

print("accuracy with scaled data:" ,accuracy_scaled)
print("accuracy with unscaled data:", accuracy_unscaled)

```

```python
#24* Write a Python program to train a Decision Tree Classifier using One-vs-Rest (OvR) strategy for multiclass
#classification*
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier

# load dataset
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data
y = iris.target

# split data into train test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state= 1)

# create and train the OneVsRestClassifier
OVR_Classifier = OneVsRestClassifier(DecisionTreeClassifier())
OVR_Classifier.fit(x_train, y_train)

# prediction
y_pred = OVR_Classifier.predict(x_test)

# accuracy calculation
accuracy = accuracy_score(y_test, y_pred)

print("accuracy with OVR Strategy:", accuracy)


```
